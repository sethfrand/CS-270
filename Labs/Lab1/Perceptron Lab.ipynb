```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# 1. Prepare the linearly separable data
X_linsep = np.array([
    [1, 2],
    [2, 3],
    [3, 1],
    [4, 2],
    [6, 6],
    [7, 6],
    [8, 5],
    [7, 6]
])
y_linsep = np.array([0, 0, 0, 0, 1, 1, 1, 1])


model = Perceptron()
model.fit(X_linsep, y_linsep)

weights = model.coef_[0]
bias = model.intercept_[0]

if weights[1] != 0:
    slope = -weights[0] / weights[1]
    intercept = -bias / weights[1]
else:
    
    slope = None 
    intercept = -bias / weights[0] 

x_min, x_max = X_linsep[:, 0].min() - 1, X_linsep[:, 0].max() + 1
y_min, y_max = X_linsep[:, 1].min() - 1, X_linsep[:, 1].max() + 1


x_line = np.linspace(x_min, x_max, 100)


if slope is not None:
    y_line = slope * x_line + intercept
else: # Case for a vertical line
    y_line = np.linspace(y_min, y_max, 100)
    x_line = np.full_like(y_line, intercept)


plt.figure(figsize=(8, 6))

plt.scatter(X_linsep[:, 0], X_linsep[:, 1], c=y_linsep, s=100, cmap='seismic', edgecolors='k')

plt.plot(x_line, y_line, 'k--', label='Decision Boundary')

plt.title('Linearly Separable Data Example')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.show()
```

What makes a dataset not linearly separable? How is the decision line determined? slope? y-intercept?

The skewness of the data is what makes a dataset not linearly seperable. This skewness is due to the relationship between features. The decision line is determine by the model after it processes the intake data and the weights, with each misclassification, the weights are adjusted. THis process is iterated a predetermined amount of epochs or until it converges, whichever comes first. The slope of the line of seperability is determined by the weights, a change in the weights is a change in the slope. The y-intercept of the line of seperability is the bias which is determined as the model runs.

```python
# Learn once on the full voting data set, while gathering info after every epoch
# Then Graph it here.  Here is some sample graph code, but you can do it your own way, including using a bar graph, etc.
import numpy as np
import matplotlib.pyplot as plt

def predict(xi, w, b):
    return 1 if np.dot(xi, w) + b >= 0 else 0

def train_perceptron_with_tracking(X, y, eta=0.1, epochs=10):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0
    misclassification_rate = []

    for epoch in range(epochs):
        errors = 0
        indices = np.arange(n_samples)
        np.random.shuffle(indices)
        
        for i in indices:
            xi, target = X[i], y[i]
            y_pred = predict(xi, w, b)
            update = eta * (target - y_pred)
            if update != 0:
                w += update * xi
                b += update
                errors += 1
        misclassification_rate.append(errors / n_samples)

    return w, b, misclassification_rate

#train 
w, b, mis_rate = train_perceptron_with_tracking(X, y, eta=0.1)

#plot
def plot_misclassification(misclassification_rate):
    plt.plot(np.arange(len(misclassification_rate)), misclassification_rate, marker="o")
    plt.title("Misclassification Rate vs. Epoch")
    plt.xlabel("Epoch")
    plt.ylabel("Misclassification Rate")
    plt.grid(True)
    plt.show()

plot_misclassification(mis_rate)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
import pandas as pd

# Different learning rates

learning_rates = [0.01, 0.1, 1.0]
results_lr = []

for eta in learning_rates:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    clf = Perceptron(eta0=eta, max_iter=1000, shuffle=True)
    clf.fit(X_train, y_train)
    
    results_lr.append({
        "eta0": eta,
        "Train Accuracy": clf.score(X_train, y_train),
        "Test Accuracy": clf.score(X_test, y_test)
    })

results_lr_df = pd.DataFrame(results_lr)
print("Results for different learning rates:")
print(results_lr_df)


# Shuffle = true vs false

results_shuffle = []

for shuf in [True, False]:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    clf = Perceptron(eta0=0.1, max_iter=1000, shuffle=shuf)
    clf.fit(X_train, y_train)
    
    results_shuffle.append({
        "Shuffle": shuf,
        "Train Accuracy": clf.score(X_train, y_train),
        "Test Accuracy": clf.score(X_test, y_test)
    })

results_shuffle_df = pd.DataFrame(results_shuffle)
print("\nResults for shuffle:")
print(results_shuffle_df)



# Fit intercept = true vs false
results_intercept = []

for intercept in [True, False]:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    clf = Perceptron(max_iter=1000, fit_intercept=intercept)
    clf.fit(X_train, y_train)
    
    results_intercept.append({
        "Fit Intercept": intercept,
        "Train Accuracy": clf.score(X_train, y_train),
        "Test Accuracy": clf.score(X_test, y_test)
    })

results_intercept_df = pd.DataFrame(results_intercept)
print("\nResults for intercept:")
print(results_intercept_df)

# Verbose

verbose_values = [0, 1, 3]
results_v = []

for v in verbose_values:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    clf = Perceptron(verbose=v, max_iter=1000)
    clf.fit(X_train, y_train)
    
    results_v.append({
        "Verbose": v,
        "Train Accuracy": clf.score(X_train, y_train),
        "Test Accuracy": clf.score(X_test, y_test)
    })

results_v_df = pd.DataFrame(results_v)
print("\nResults for different verbose levels:")
print(results_v_df)

# random states

random_states = [1, 43, 7]
results_random = []

for rs in random_states:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=rs)
    
    clf = Perceptron(max_iter=1000)
    clf.fit(X_train, y_train)
    
    results_random.append({
        "Random State": rs,
        "Train Accuracy": clf.score(X_train, y_train),
        "Test Accuracy": clf.score(X_test, y_test)
    })

results_random_df = pd.DataFrame(results_random)
print("\nResults for different random states:")
print(results_random_df)
```
